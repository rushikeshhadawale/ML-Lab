```python
from sklearn.cluster import KMeans 
import pandas as pd
import matplotlib.pyplot as plt 
from matplotlib import style 

style.use("ggplot")
data = pd.DataFrame([
    [1, 2], [5, 8], [1.5, 1.8],
    [8, 8], [1, 0.6], [9, 11]
], columns=['x','y'])

print("Dataset:")
print(data)

kmeans = KMeans(n_clusters=2, random_state=42).fit(data)

centroids = kmeans.cluster_centers_
labels = kmeans.labels_

print("\nCluster Centroids:")
print(centroids)
print("\nLabels:")
print(labels)

plt.scatter(data['x'], data['y'], c=labels, cmap='rainbow')
plt.scatter(centroids[:,0], centroids[:,1], c='black', marker='X', s=200)
plt.title("KMeans Clustering Example")
plt.show()
```
